```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from numpy import set_printoptions
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
from sklearn.cluster import DBSCAN
import warnings
warnings.filterwarnings("ignore")
```

## Hierarchical Clustering

```python
data=pd.read_csv("Crime_data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.info()
```

```python
# Normalizing the data
```

```python
def norm_fun(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
data_norm=norm_fun(data.iloc[:,1:])
```

```python
data_norm
```

```python
plt.figure(figsize=(20,5))
den=sch.dendrogram(sch.linkage(data_norm, method='single'))
```

```python
plt.figure(figsize=(20,5))
den=sch.dendrogram(sch.linkage(data_norm, method='complete'))
```

```python
plt.figure(figsize=(20,5))
den=sch.dendrogram(sch.linkage(data_norm, method='average'))
```

```python
plt.figure(figsize=(20,5))
den=sch.dendrogram(sch.linkage(data_norm, method='centroid'))
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
```

```python
hc
```

```python
y_hc=hc.fit_predict(data_norm)
clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
y_hc
```

```python
data1=data.copy()
```

```python
data1['Clusterid']=hc.labels_
```

```python
data1
```

```python
data1.iloc[:,1:].groupby(data1.Clusterid).mean()
```

## Kmeans Clustering

### Above as we had normalized data we can use that and continue

```python
data_norm
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
cluster_new=KMeans(n_clusters=4)
cluster_new.fit(data_norm)
cluster_new.labels_
```

```python
data2=data.copy()
```

```python
data2.head()
```

```python
data2['ClusteridNew']=cluster_new.labels_
```

```python
data2
```

```python
data2.iloc[:,1:].groupby(data2.ClusteridNew).mean()
```

## DBSCAN Clustering

```python
data3=data.copy()
```

```python
data3
```

```python
data3=data3.iloc[:,1:]
```

```python
data3
```

```python
X=data3.values
```

```python
scaler=StandardScaler().fit(X)
X1=scaler.transform(X)
```

```python
X1
```

```python
dbscan=DBSCAN(eps=0.8, min_samples=4)
dbscan.fit(X1)
```

```python
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_, columns=['Cluster'])
```

```python
pd.concat([data3, c1],axis=1)
```
