```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Wine.csv")
```

```python
data.head()
```

```python
wine=data.copy()
```

```python
wine
```

```python
wine1=wine.iloc[:,1:]
```

```python
wine.head()
```

```python
wine2=wine1.values
```

```python
wine2
```

```python
wine3=scale(wine2)
```

```python
wine3
```

```python
pca=PCA(n_components=13)
pca_values=pca.fit_transform(wine3)
```

```python
pca_values
```

```python
var=pca.explained_variance_ratio_
```

```python
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

#### As we can see the most of variance is put in the first 3 components. Since there is not much variance present from 4th component, lets just consider the first 3 componets in our analysis.

## Hierarchical Clustering

```python
data1=data.copy()
```

```python
data2=data1.iloc[:,:3]
```

```python
data2
```

```python
def norm_fun(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
data2_norm=norm_fun(data2.iloc[:,:])
```

```python
data2_norm
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
```

```python
hc
```

```python
y_hc=hc.fit_predict(data2_norm)
clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
y_hc
```

```python
datahc=data2.copy()
```

```python
datahc['Cluster']=hc.labels_
```

```python
datahc
```

```python
datahc.iloc[:,:].groupby(datahc.Cluster).mean()
```

## Kmeans Clustering

```python
data2_norm
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(data2_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,10),wcss)
plt.show
    
```

```python
cluster_new=KMeans(n_clusters=3)
cluster_new.fit(data2_norm)
cluster_new.labels_
```

```python
datakm=data2.copy()
```

```python
datakm['Cluster']=cluster_new.labels_
```

```python
datakm
```

```python
datakm.iloc[:,:].groupby(datakm.Cluster).mean()
```
