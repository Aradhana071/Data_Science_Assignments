```python
userdata=data1.pivot_table(index='userid',
                    columns='booktitle',
                    values='bookrating').reset_index(drop=True)
```

```python
userdata
```

```python
userdata.fillna(0, inplace=True)
```

```python
userdata
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( userdata.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = data1.userid.unique()
user_sim_df.columns = data1.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```
